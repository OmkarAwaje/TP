```python
class N_Queens:
    def __init__(self,n):
        self.n=n
    def is_safe(self,board,row,col):
        for i in range(self.n):
            if(board[row][i]=="Q"):
                return False
        for i in range(self.n):
            if(board[i][col]=="Q"):
                return False
        r=row
        for i in range(col,self.n):
            if(r>=0):
                if(board[r][i]=="Q"):
                    return False
            r-=1
        c=col
        for i in range(row,self.n):
            if(c>=0):
                if(board[i][c]=="Q"):
                    return False
            c-=1
        r=row
        c=col
        while(r>=0 and c>=0):
            if(board[r][c]=="Q"):
                return False
            r-=1
            c-=1
        r=row
        c=col
        while(r<self.n and c<self.n):
            if(board[r][c]=="Q"):
                return False
            r+=1
            c+=1
        return True
    def solve(self,board,ans,col):
        if col==self.n:
            ans.append(["".join(row) for row in board])
            return
        for row in range(self.n):
            if board[row][col]=="." and self.is_safe(board,row,col):
                board[row][col]="Q"
                self.solve(board,ans,col+1)
                board[row][col]="."
            if(board[row][col]=="Q"):
                self.solve(board,ans,col+1)
ans=[]
n=int(input("Enter number of Queens:"))
x=int(input("Enter row to place first queen:"))
y=int(input("Enter column of to place first queen:"))
x=x-1
y=y-1
board=[["." for _ in range(n)] for _ in range(n)]
board[x][y] = "Q"
a=N_Queens(n)
a.solve(board,ans,0)

print("Possibilities: ",len(ans))
if(len(ans)>0):
    for solution in ans:
        print("Solution:")
        for row in solution:
            print(row)
        print()

            
            
                
        
```
